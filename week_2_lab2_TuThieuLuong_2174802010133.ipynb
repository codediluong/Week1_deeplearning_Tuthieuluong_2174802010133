Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT

```python
import pandas as pd
import numpy as np

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

pivot = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV"],
    aggfunc=[
        "count",
        "sum",
        "mean",
        "median",
        "min",
        "max",
        "std",
        lambda x: x.quantile(0.25),  # Q1
        lambda x: x.quantile(0.50),  # Q2
        lambda x: x.quantile(0.75)   # Q3
    ]
)

# Đổi tên cho dễ đọc
pivot.columns = [
    "count", "sum", "mean", "median", "min", "max", "std",
    "Q1", "Q2", "Q3"
]

pivot
```

5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT, KV và DT

```python
import pandas as pd
import numpy as np

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

pivot = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV", "DT"],
    aggfunc=[
        "count",
        "sum",
        "mean",
        "median",
        "min",
        "max",
        "std",
        lambda x: x.quantile(0.25),  # Q1
        lambda x: x.quantile(0.50),  # Q2
        lambda x: x.quantile(0.75)   # Q3
    ]
)

# Đặt lại tên cột cho rõ ràng
pivot.columns = [
    "count", "sum", "mean", "median", "min", "max", "std",
    "Q1", "Q2", "Q3"
]

pivot
```

Phần 2: Trình bày dữ liệu

```python
# Bảng tần số & tần suất
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

freq_GT = df["GT"].value_counts()
ratio_GT = df["GT"].value_counts(normalize=True)

bang_GT = pd.DataFrame({
    "Tần số": freq_GT,
    "Tần suất": ratio_GT
})

bang_GT

# Biểu đồ tần số (cột) - Biểu đồ tần số (tròn)

import matplotlib.pyplot as plt

freq_GT.plot(kind="bar")
plt.xlabel("Giới tính")
plt.ylabel("Tần số")
plt.title("Biểu đồ tần số giới tính")
plt.show()

ratio_GT.plot(kind="pie", autopct="%.1f%%")
plt.ylabel("")
plt.title("Biểu đồ tần suất giới tính")
plt.show()

# 2. Trình bày dữ liệu các biến US_TBM1, US_TBM2, US_TBM3
df[["US_TBM1","US_TBM2","US_TBM3"]].describe()
df

# 3. Trình bày dữ liệu biến DT với các học sinh là nam
df_nam = df[df["GT"] == "Nam"]
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

freq_DT = df_nam["DT"].value_counts()
ratio_DT = df_nam["DT"].value_counts(normalize=True)

pd.DataFrame({
    "Tần số": freq_DT,
    "Tần suất": ratio_DT
})

# 4. Trình bày dữ liệu biến KV với học sinh thỏa điều kiện
# Giới tính: Nam
# Dân tộc: Kinh
# DH1 ≥ 5.0, DH2 ≥ 4.0, DH3 ≥ 4.0

dk = (
    (df["GT"] == "Nam") &
    (df["DT"] == "Kinh") &
    (df["DH1"] >= 5.0) &
    (df["DH2"] >= 4.0) &
    (df["DH3"] >= 4.0)
)

df_loc = df[dk]
dk

freq_KV = df_loc["KV"].value_counts()
ratio_KV = df_loc["KV"].value_counts(normalize=True)

pd.DataFrame({
    "Tần số": freq_KV,
    "Tần suất": ratio_KV
})

# 5. Trình bày dữ liệu các biến DH1, DH2, DH3 ≥ 5.0 và thuộc khu vực 2NT

df_2nt = df[
    (df["KV"] == "2NT") &
    (df["DH1"] >= 5.0) &
    (df["DH2"] >= 5.0) &
    (df["DH3"] >= 5.0)
]
df_2nt[["DH1", "DH2", "DH3"]].describe()




```

Phần 3: Trực quan hóa dữ liệu theo nhóm phân loại


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# 1. Trực quan học sinh NỮ theo XL1, XL2, XL3 (dạng unstacked)
df_nu = df[df["GT"] == "Nữ"]

xl = (
    df_nu
    .groupby(["XL1","XL2","XL3"])
    .size()
    .unstack(fill_value=0)
)
xl.plot(kind="bar")
plt.xlabel("Xếp loại lớp 10 (XL1)")
plt.ylabel("Số lượng học sinh nữ")
plt.title("Học sinh nữ theo XL1, XL2, XL3")
plt.show()
```
