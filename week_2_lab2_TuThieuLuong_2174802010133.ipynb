Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT

```python
import pandas as pd
import numpy as np

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

pivot = pd.pivot_table(
    df,
    values="DH1",
    index=["KT", "KV"],
    aggfunc=[
        "count",
        "sum",
        "mean",
        "median",
        "min",
        "max",
        "std",
        lambda x: x.quantile(0.25),  # Q1
        lambda x: x.quantile(0.50),  # Q2
        lambda x: x.quantile(0.75)   # Q3
    ]
)

# Đổi tên cho dễ đọc
pivot.columns = [
    "count", "sum", "mean", "median", "min", "max", "std",
    "Q1", "Q2", "Q3"
]

pivot
```

Phần 3: Trực quan hóa dữ liệu theo nhóm phân loại


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# 1. Trực quan học sinh NỮ theo XL1, XL2, XL3 (dạng unstacked)
df_nu = df[df["GT"] == "Nữ"]

xl = (
    df_nu
    .groupby(["XL1","XL2","XL3"])
    .size()
    .unstack(fill_value=5)
)
df.plot(kind="bar")
plt.xlabel("Xếp loại lớp 10 (XL1)")
plt.ylabel("Số lượng học sinh nữ")
plt.title("Học sinh nữ theo XL1, XL2, XL3")
plt.show()

# 2. Trực quan KQXT với khối A, A1, B thuộc khu vực 1, 2
df_kq = df[
    (df["KT"].isin(["A","A1","B"])) &
    (df["KV"].isin(["1","2"]))
]
df_kq.groupby(["KT","KQXT"]).size().unstack().plot(kind="bar")
plt.xlabel("Khối thi")
plt.ylabel("Số lượng thí sinh")
plt.title("KQXT theo khối A, A1, B (KV 1,2)")
plt.show()

# 3. Trực quan số lượng thí sinh từng khu vực theo từng khối thi
df.groupby(["KV","KT"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Khu vực")
plt.ylabel("Số lượng thí sinh")
plt.title("Số lượng thí sinh theo KV và KT")
plt.show()

# 4. Trực quan số lượng thí sinh ĐẬU / RỚT theo từng khối thi
df.groupby(["KT","KQXT"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Khối thi")
plt.ylabel("Số lượng thí sinh")
plt.title("Đậu – Rớt theo khối thi")
plt.show()

# 5. Trực quan số lượng thí sinh ĐẬU / RỚT theo từng khu vực
df.groupby(["KV","KQXT"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Khu vực")
plt.ylabel("Số lượng thí sinh")
plt.title("Đậu – Rớt theo khu vực")
plt.show()

# 6. Trực quan số lượng thí sinh ĐẬU / RỚT theo từng dân tộc
df.groupby(["DT","KQXT"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Dân tộc")
plt.ylabel("Số lượng thí sinh")
plt.title("Đậu – Rớt theo dân tộc")
plt.show()

# 7. Trực quan số lượng thí sinh ĐẬU / RỚT theo từng giới tính
df.groupby(["GT","KQXT"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Giới tính")
plt.ylabel("Số lượng thí sinh")
plt.title("Đậu – Rớt theo giới tính")
plt.show()


```

PHẦN 4: TRỰC QUAN HÓA DỮ LIỆU NÂNG CAO

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# 1. Vẽ biểu đồ đường (Simple Line) cho biến T1
t1_count = df["T1"].value_counts().sort_index()
plt.plot(t1_count.index, t1_count.values, marker="o")
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ đường Simple Line cho biến T1")
plt.grid(True)
plt.show()

# 2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
# a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
# b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
# c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
# d. Từ 8 trở lên = giỏi (ký hiệu “g”)

def phanloai_t1(x):
    if x < 5:
        return "Kém"
    elif x < 7:
        return "Trung bình"
    elif x < 8:
        return "Khá"
    else:
        return "Giỏi"

df["phanlopT1"] = df["T1"].apply(phanloai_t1)

# 3. Lập bảng tần số cho biến phanloait1
freq_phanlopT1 = df["phanlopT1"].value_counts()
freq_phanlopT1

#4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
multi = (
    df.groupby(["phanlopT1", "T1"])
    .size()
    .unstack(fill_value=0)
)
for label in multi.index:
    plt.plot(multi.columns, multi.loc[label], marker="o", label=label)

plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ Multiple Line cho T1 theo phân loại")
plt.legend(title="phanlopT1")
plt.grid(True)
plt.show()

# 5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1
plt.figure(figsize=(10,6))

for label in multi.index:
    x = multi.columns
    y = multi.loc[label]
    plt.vlines(x, ymin=0, ymax=y)
    plt.plot(x, y, 'o', label=label)

plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ Drop-line cho T1 theo phân loại")
plt.legend(title="phanlopT1")
plt.show()

# PHẦN 5: MÔ TẢ DỮ LIỆU VÀ KHẢO SÁT DẠNG PHÂN PHỐI
# 1a) Mô tả độ tập trung và phân tán (T1)
t1 = df["T1"].dropna()

summary = t1.describe()  # count, mean, std, min, Q1, median, Q3, max
summary
skew = t1.skew()
kurt = t1.kurt()  # excess kurtosis
skew, kurt

# Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
Q1 = t1.quantile(0.25)
Q2 = t1.quantile(0.50)
Q3 = t1.quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR
min_val = t1.min()
max_val = t1.max()

out_low = t1[t1 < lower_fence]
out_high = t1[t1 > upper_fence]

ten_dai_luong = {
    "Min": min_val,
    "Q1": Q1,
    "Median(Q2)": Q2,
    "Q3": Q3,
    "Max": max_val,
    "IQR": IQR,
    "Lower fence": lower_fence,
    "Upper fence": upper_fence,
    "Outliers thấp (count)": int(out_low.shape[0]),
    "Outliers cao (count)": int(out_high.shape[0]),
}
ten_dai_luong

plt.boxplot(t1, vert=True)
plt.ylabel("T1")
plt.title("Boxplot của T1")
plt.show()

# 1c) Histogram (hình dạng phân phối)
plt.hist(t1, bins=10)
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.title("Histogram của T1")
plt.show()

# Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn 

```
